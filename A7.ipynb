```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# ==========================================
# Step 1: Load High-Dimensional Data
# ==========================================
# Breast Cancer dataset has 30 features (dimensions)
data = load_breast_cancer()
X = data.data
y = data.target # We won't use this for training, only for checking later if we want

print(f"Original shape: {X.shape}") # Shows (569 rows, 30 columns)

# ==========================================
# Step 2: Scaling (CRITICAL for PCA)
# ==========================================
# PCA is sensitive to scale. We must normalize data first.
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# ==========================================
# Step 3: Apply PCA (Dimensionality Reduction)
# ==========================================
# We reduce 30 columns down to just 2 Principal Components
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

print(f"Reduced shape: {X_pca.shape}") # Shows (569 rows, 2 columns)

# ==========================================
# Step 4: Apply K-Means Clustering
# ==========================================
# We cluster based on the new 2D data
kmeans = KMeans(n_clusters=2, random_state=42)
labels = kmeans.fit_predict(X_pca)

# ==========================================
# Step 5: Evaluate Quality
# ==========================================
score = silhouette_score(X_pca, labels)
print(f"Silhouette Score: {score:.2f}")

# ==========================================
# Step 6: Visualization
# ==========================================


[Image of PCA Plot]

plt.figure(figsize=(8, 6))

# Scatter plot of the two Principal Components
# c=labels colors the points based on the cluster K-Means found
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels, cmap='viridis', edgecolor='k')

plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA + K-Means Clustering Results')
plt.show()
```
